```python
import pandas as pd
import altair as alt
```

```python
dataset = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
all_countries = dataset.groupby(['location', 'date'])['new_cases_smoothed_per_million'].mean().reset_index()
filter_countries = ['Spain', 'France', 'Brazil']
df = all_countries[all_countries['location'].isin(filter_countries)]
```

```python
plot = alt.Chart(df).mark_area(opacity=0.4, stroke='black').encode(
    x='date:T',
    y=alt.Y('new_cases_smoothed_per_million:Q', stack=None),
    color=alt.Color('location:N', scale=alt.Scale(scheme='set1')),
    tooltip='location'
).interactive().properties(width='container')
plot
```

```python
import datapane as dp
dp.Report(dp.Plot(plot), dp.Table(df)).publish(name='covid_country_comparison', open=True)
```
